## Expungement Data

```python
expunge = pd.read_csv("jonData.csv").dropna(subset=['fips'])
```

```python
# first to int
expunge['fips'] = expunge['fips'].astype(int)
# then to str
expunge['fips'] = expunge['fips'].astype(str)

# then add the leading zeros
expunge['fips'] = expunge['fips'].apply(lambda x: '{0:0>4}'.format(x))
expunge
```

# Now merge compDF (now called 'compDF_merge') and expunge

```python
final = compDF_merge.merge(expunge, how='inner', on='fips')
final
```
